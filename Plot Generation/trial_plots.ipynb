```python
session_0 = pd.read_csv('/Users/varshinisubramanian/Desktop/CMU/Yttri Lab Research/No Conflict Low Volatility (control)/CSV Files/session_0_mouse9.csv')
session_0_states = pd.DataFrame(session_0[['index', 'state']])
```

```python
session_0_full = pd.merge(session_0_full, session_0_states, on='index', how='left')
```

```python
# lines: 
# 1 - correct action, 0 - incorrect action
# long line - reward, short line - unrewarded
# blue - left turn, green - right turn
# black dotted lines - block changes

fig, ax = plt.subplots(figsize=(30, 10))
#ax.xlim([0, len(session_0_full)])
ax2 = ax.twinx()
ax2.set_ylim([-2, 6.5])
ax.set_ylabel('Correct Action Taken (Binary)')
ax2.set_ylabel('States')

for index, row in session_0_full.iterrows():
    if row['event_marker'] == 1: ax.vlines(index, 0.90, 1)
    elif row['event_marker'] == 5: ax.vlines(index, 0.90, 1, color='pink')
    elif row['event_marker'] == 2: ax.vlines(index, 0.95, 1)
    elif row['event_marker'] == 6: ax.vlines(index, 0.95, 1, color='pink')
    elif row['event_marker'] == 3: ax.vlines(index, 0, 0.10)
    elif row['event_marker'] == 7: ax.vlines(index, 0, 0.10, color='pink')
    elif row['event_marker'] == 4: ax.vlines(index, 0, 0.05)
    elif row['event_marker'] == 8: ax.vlines(index, 0, 0.05, color='pink')

    if row['block_change'] == 1: ax.vlines(index, 0, 1.05, color='black', linestyles='dotted')

    if row['block'] == 'right': ax.hlines(1.03, index, index + 1, color='pink', linewidth=13.)
    elif row['block'] == 'left': ax.hlines(1.03, index, index + 1, linewidth=13.)

    if row['state'] == 0: ax2.vlines(index, -0.25, 0.25, color='red')
    elif row['state'] == 1: ax2.vlines(index, 0.75, 1.25, color='orange')
    elif row['state'] == 2: ax2.vlines(index, 1.75, 2.25, color='blue')
    elif row['state'] == 3: ax2.vlines(index, 2.75, 3.25, color='black')
    elif row['state'] == 4: ax2.vlines(index, 3.75, 4.25, color='m')
    elif row['state'] == 5: ax2.vlines(index, 4.75, 5.25, color='olive')
    elif row['state'] == 6: ax2.vlines(index, 5.75, 6.25, color='brown')
    else: ax2.vlines(index, -0.25, 0.25, color='green')

plt.show()
```
